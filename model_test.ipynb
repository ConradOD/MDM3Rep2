```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Some variables
l = 10 #size of square grid
N = 4 #number of planes
Nt = 10 #number of timesteps
dt = 0.5 #size of timesteps
max_speed = 1
```

```python
#Random plane initial positions
x = np.zeros([N,Nt])
y = np.zeros([N,Nt])
x[:,0] = np.random.rand(N)*l
y[:,0] = np.random.rand(N)*l
```

```python
#Random plane initial velocity / direction
theta = 2 * np.pi * np.random.rand(1,N)
speed = np.random.rand(1,N) * max_speed
vx = speed * np.cos(theta)
vy = speed * np.sin(theta)
```

```python
#Calculate distance between plane and all others
#Returns minimum distance
def get_min_dist(i):
    #i is index of plane
    for index in np.concatenate((np.arange(0,i),np.arange(i+1,N))):
        print(index)

get_min_dist(2)
```

```python
for i in range(1,Nt):
    #Move
    x[:,i] = x[:,i-1] + vx * dt
    y[:,i] = y[:,i-1] + vy * dt

    #Make sure plane in grid (very unrealistic, for simplicity)
    x[:,i] = x[:,i]%l
    y[:,i] = y[:,i]%l

    
```

```python
plt.figure()
for i in range(N):
    plt.scatter(x[i,0],y[i,0])
    plt.plot(x[i,:],y[i,:],label=f'Plane {i}')

plt.xlim([0,l])
plt.ylim([0,l])
plt.legend()
```

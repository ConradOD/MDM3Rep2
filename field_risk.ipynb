```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)
```

```python
f = np.zeros((x.size,y.size))
for i in range(x.size):
    for j in range(y.size):
        f[i,j]  = 1/(x[i]**2 + y[j]**2 + 5)
```

```python
plt.imshow(f)
plt.colorbar()
```
